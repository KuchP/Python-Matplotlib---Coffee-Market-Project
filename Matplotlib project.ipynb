# COMPOSITION OVER TIME

```python
comparision.head()
```

# COMPOSITION OF COFFEE PRODUCTION

* Create a Pie chart with the top 5 producers, together with the sum for the rest of countries.

* Finally, create two donut charts, one representing Brazil's share of global production in 1990, and one for Brazil's share in 2018.

```python
coffee_production["Rest of World"] = coffee_production.drop(["Brazil", 
                                                             "Viet Nam", 
                                                             "Colombia", 
                                                             "Indonesia",
                                                             "Ethiopia",], axis=1).sum(axis=1)


coffee_production_top5_2018 = coffee_production.loc['2018', ["Brazil", 
                                                             "Viet Nam", 
                                                             "Colombia", 
                                                             "Indonesia",
                                                             "Ethiopia",
                                                             "Rest of World"
                                                            ]
                                                   ]

coffee_production_top5_2018
```

```python
df_gr = coffee_production_top5_2018.sum()
```

```python
df_gr
```

```python
fig, ax = plt.subplots(figsize = (10,10))

ax.pie(df_gr[::-1],
       autopct="%.0f%%",
       labels=list(df_gr.index[::-1]),
       startangle=90,
       explode = (0,0,0,0,0,0.2)
      )

fig.suptitle("Share of Global Coffee Production 2018", fontsize = 25, y = 0.90)

plt.show()
```

```python
brazil_vs_others = (coffee_production
                    .assign(rest_of_world = coffee_production
                            .drop("Brazil", axis=1).sum(axis=1))
                    .loc[:, ["Brazil", "rest_of_world"]]
                    .astype({"Brazil": "float64"})
                   )


brazil_vs_others.head()
```

```python
fig, ax = plt.subplots()

ax.pie(
    brazil_vs_others.iloc[0].sort_values(ascending=False),
    startangle=90,
    labels = ["", ""],
    #autopct='%.0f%%',
    pctdistance=.85,
    colors=["White", "Blue"]
)

hole = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
  
# Adding Circle in Pie chart
fig.gca().add_artist(hole)

plt.text(
    0, 
    0, 
    f"{round(brazil_vs_others.loc['1990', 'Brazil'].sum() / brazil_vs_others.loc['1990'].sum().sum()*100)}%", 
    ha='center', 
    va='center', 
    fontsize=42
)
  

ax.set_title("Brazil Share of Global Production 1990", fontsize=14)

plt.show()
```

```python
fig, ax = plt.subplots()

ax.pie(
    brazil_vs_others.iloc[0].sort_values(ascending=False),
    startangle=90,
    labels = ["", ""],
    #autopct='%.0f%%',
    pctdistance=.85,
    colors=["White", "Blue"]
)

hole = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
  
# Adding Circle in Pie chart
fig.gca().add_artist(hole)

plt.text(
    0, 
    0, 
    f"{round((brazil_vs_others.loc['2018', 'Brazil'].sum() / brazil_vs_others.loc['2018'].sum().sum()*100))}%", 
    ha='center', 
    va='center', 
    fontsize=42
)
  

ax.set_title("Brazil Share of Global Production 2018", fontsize=14)

plt.show()
```

# Coffee Consumption Vs. Prices

Compare coffee prices with coffee consumption in the following countries. 

Plot a dual axis bar chart of price vs consumption totals for these countries.

```python
consumption = pd.read_csv(
    r"\Users\pkucharzak\Desktop\PYTHON FILE\Matplotlib and Seaborn Python\Data Viz with Matplotlib & Seaborn\Data\coffee\imports.csv",
).set_index('imports').mean(axis = 1)


consumption.head()
```

```python
prices = pd.read_csv(
    r"\Users\pkucharzak\Desktop\PYTHON FILE\Matplotlib and Seaborn Python\Data Viz with Matplotlib & Seaborn\Data\coffee\retail-prices.csv"
).set_index('retail_prices').mean(axis = 1)
```

```python
prices
```

```python
price_cons = pd.DataFrame(consumption).merge(
    pd.DataFrame(prices),
    how = 'inner',
    left_on = consumption.index,
    right_on = prices.index
)
```

```python
price_cons.columns = ['countries','consumption','prices']
price_cons['countries'] = price_cons['countries'].str.strip(" ")
```

```python
price_cons = price_cons.sort_values('consumption', ascending = False)
```

```python
price_cons
```

```python
fig, ax = plt.subplots(figsize = (15,10))

width = .25
x=np.arange(0, len(price_cons["countries"]))

bar1 = ax.bar(
    x-width/2,
    price_cons['consumption'],
    width = width,
    label = 'consumption'
)

ax.set_ylabel('Consumption in Thousands (Bags 60kg)', fontsize = 12)

ax2 = ax.twinx()

bar2 = ax2.bar(
    x+width/2,
    price_cons['prices'],
    width = width,
    label = 'prices',
    color = 'orange'
)
ax2.set_ylabel('Prices in hundreds (Bags 60kg)', fontsize = 12)


plt.xticks(
    x,
    price_cons['countries']
    
)

fig.suptitle('Comparision between consumption vs Prices since 1990 - 2018', fontsize = 20, y = .92)
plt.legend([bar1,bar2],['Consumption','Prices'])
plt.show()
```
